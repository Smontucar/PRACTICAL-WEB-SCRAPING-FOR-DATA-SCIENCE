```python
# Bucles infinitos
# ¿Olvidó agregar la línea x = x + 1 ? Probado while True: ¿ o su bucle "para" va en círculos? 
# Puede presionar Control + C en su teclado para ejecutar una “Interrupción de teclado” y detener la ejecución del 
# bloque de código actual.

# Siempre que esté escribiendo código, es una buena idea encapsular pequeños fragmentos de código reutilizables para que 
# puedan ser llamados y ejecutados en varios lugares sin tener que copiar y pegar mucho código. Una forma básica de 
# hacerlo es crear una función, que se hace usando "def":
```

```python
def add(x, y):
    print("x is {} and y is {}".format(x, y))
    return x + y # return value
```

```python
add(5, 6)
```

```python
add(y=10, x=5)
```

```python
# Hay dos construcciones especiales que vale la pena mencionar aquí también: “*” y “**”. Ambos se pueden usar en 
# firmas de funciones para indicar "el resto de los argumentos" y "el resto de los argumentos nombrados" respectivamente. 
# Demostremos cómo funcionan con un ejemplo
```

```python
def many_arguments(*args):
    # args will be a tuple / será una tupla
    print(args)
```

```python
many_arguments(1, 2, 3)
```

```python
def many_named_arguments(**kwargs):
    # kwargs will be a dictionary / será un diccionario
    print(kwargs)
```

```python
many_named_arguments(a=1, b=2)
```

```python
def both_together(*args, **kwargs):
    print(args, kwargs)
```

```python
# Además de usarlos en firmas de métodos, también puede usarlos al llamar a funciones para indicar que desea pasar 
# un iterable como argumentos o un diccionario como argumentos con nombre
```

```python
def add(a, b):
    return a + b
```

```python
l = [1,2] # tuples work, too
add(*l)
```

```python
d = {'a': 2, 'b': 1}
add(**d)
```
